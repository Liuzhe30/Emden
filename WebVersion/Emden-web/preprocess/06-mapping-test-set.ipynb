### mapping test set (to drug and gene)

```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
feature_dataset = pd.read_pickle('../datasets/dataset_featurecode.dataset')
print(feature_dataset)
```

```python
testset = pd.read_pickle('../datasets/middlefile/test_data_evidence.dataset')
print(testset)
```

```python
import numpy as np
test_pred_result = np.load('../model/pred_results/ori/pred.npy')
```

```python
import warnings
warnings.filterwarnings('ignore')
from tqdm import tqdm
full_testset = pd.DataFrame(columns=['gene', 'variant', 'drug', 'source', 'label', 'pred_result'])
for i in tqdm(range(testset.shape[0])): # range(testset.shape[0])
    smile = testset['smile'][i]
    label = testset['label'][i]
    hhmb_first = testset['variantfeature'][i][0]
    hhma_first = testset['variantfeature'][i][1830]
    pred_result = test_pred_result[i]
    for j in range(feature_dataset.shape[0]):
        if(smile == feature_dataset['smile'][j] and label == feature_dataset['label'][j] and hhmb_first == feature_dataset['hhm_before'][j][0][0] and hhma_first == feature_dataset['hhm_after'][j][0][0]):
            gene = feature_dataset['gene'][j]
            variant = feature_dataset['variant'][j]
            drug = feature_dataset['drug'][j]
            source = feature_dataset['source'][j]
            full_testset = full_testset.append([{'gene': gene,  'variant': variant, 'drug': drug,  
                                        'source':source, 'label': label, 'pred_result': pred_result}], ignore_index=True)
            break
print(full_testset)
full_testset.to_csv('../datasets/testset.csv', index=None)
```
